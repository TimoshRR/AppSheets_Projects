```
!pip install transformers

from transformers import pipeline

# Assuming 'text' holds the context for which you want to generate questions,
# you'll need to structure it into a suitable input for the QA model.

# Example:
text = "Extracted text undergoes preprocessing steps, including tokenization, stopword removal, and stemming/lemmatization, using popular NLP libraries like NLTK or SpaCy. This process ensures the text is clean and ready for further analysis and question formation.For generating questions, transformer-based models like BERT or T5 will be employed. These models are trained on large datasets of question-answer pairs, enabling them to generate contextually relevant questions based on input text."
# Changed 'question-generation' to 'question-answering' and specified the model.
# Question Generation Pipeline
qa_generator = pipeline("text2text-generation", model="valhalla/t5-base-qa-qg-hl")

#  Prompt the model to generate questions and answers
qa_result = qa_generator(f"generate questions and answers: {text}")

# Extract and print questions and answers
for item in qa_result:
    generated_text = item['generated_text']
    if "Question:" in generated_text and "Answer:" in generated_text:
        question = generated_text.split("Question:")[1].split("Answer:")[0].strip()
        answer = generated_text.split("Answer:")[1].strip()
        print(f"Question: {question}")
        print(f"Answer: {answer}\n")
```

```
!pip install transformers

from transformers import pipeline

text = "Python is an interpreted, object-oriented, high-level programming language with dynamic semantics."

# Use a text-to-text generation pipeline for both questions and answers
qa_generator = pipeline("text2text-generation", model="valhalla/t5-base-qa-qg-hl")

# Prompt the model to generate questions and answers
qa_result = qa_generator(f"generate questions and answers: {text}")

# Extract and print questions and answers using a more robust approach
for item in qa_result:
    generated_text = item['generated_text']
    # Split by lines and look for question/answer patterns
    lines = generated_text.split('\n')
    for i in range(len(lines)):
        if lines[i].startswith("Question:"):
            question = lines[i].replace("Question:", "").strip()
            if i + 1 < len(lines) and lines[i + 1].startswith("Answer:"):
                answer = lines[i + 1].replace("Answer:", "").strip()
                print(f"Question: {question}")
                print(f"Answer: {answer}\n")
```

```
# Example usage:
result = nlp(question="What is Python?", context=text)
print(result)
```

```
from transformers import pipeline

# Set up the question generation pipeline
text = "Facewashes have the tendency to cleanse the dirt particles and make your skin look fresh by purifying it. But using harsh facewashes can also lead to the damage of your skin. Its better to strictly avoid products containing Aluminium, Mineral oil, any kind of Parabens, Polyethylene glycol, Sodium Laurel sulphate, Sodium Laureth Sulphate and Synthetic fragments. Gentle facewash can help your skin be clean and healthy."
nlp_qg = pipeline("text2text-generation", model="valhalla/t5-base-qa-qg-hl")

# Set up the question-answering pipeline
questions = nlp_qg(
    f"generate questions: {text}",
    num_return_sequences=3,  # Number of questions to generate per run
    do_sample=True,          # Enable sampling to introduce randomness
    top_k=50                  # Control diversity in sampling
)

# Set up the question-answering pipeline
nlp_qa = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Generate questions and answers
for item in questions:
    question = item['generated_text']
    answer = nlp_qa(question=question, context=text)
    print(f"Q: {question}")
    print(f"A: {answer['answer']}\n")
```
